This is a bar chart of a couple of USA, India, Brazil and South Africa using health data set from Tableau.com

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
TB_Burden_Country = pd.read_csv('https://public.tableau.com/app/sample-data/TB_Burden_Country.csv?_gl=1*1p865mv*_ga*MTQyNzExMTAxMy4xNzc5NDc2MDcy*_ga_8YLN0SNXVS*czE3Nzk2NDkxNzMkbzMkZzAkdDE3Nzk2NDkxNzMkajYwJGwwJGgw*_gcl_au*MTQ2MzMyMTcxOC4xNzc5NDc2MDcy')
```

```python
df = TB_Burden_Country
col = 'Estimated mortality of TB cases (all forms, excluding HIV), per 100 000 population, high bound'
```

```python
year = 2013
data = df[df['Year'] == year]
countries = ['Brazil', 'India','South Africa']
subset = data[data['Country or territory name'].isin(countries)]


plt.bar(subset['Country or territory name'], subset[col], color=['navy', 'orange', 'green', 'crimson'])
plt.ylabel('Mortality rates 2013 per 100,000)')

plt.title(f'TB Mortality Rates in {year}')
plt.xticks(rotation=45, ha ='right')
plt.show()
#
```
